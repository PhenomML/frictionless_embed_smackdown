# Paper Artifact Generation

Front end notebook for our paper artifact pipeline. Analysis and plotting logic lives in `benchmark.paper_analysis` and `benchmark.paper_figures`; the CLI entrypoint is `scripts/generate_paper_figures.py`.


````python
from pathlib import Path
import subprocess
import sys

root = Path.cwd()
if root.name == "notebooks":
    root = root.parent

cmd = [
    sys.executable,
    "scripts/generate_paper_figures.py",
    "--config",
    "configs/paper_reproduction.json",
]

completed = subprocess.run(cmd, cwd=root, check=True, text=True, capture_output=True)
print(completed.stdout)
````

## CLI Equivalent

Run the same pipeline outside notebooks with:

```bash
python scripts/generate_paper_figures.py --config configs/paper_reproduction.json
```
